### Ans 1:-

```python
import numpy as np
import matplotlib.pyplot as plt
A = 1
tau = 1
t = np.linspace(0,2,100)
W = A*(np.exp( -t/tau) - 1 + t/tau)
fig = plt.figure()
plt.plot(t,W, 'r')
plt.xlabel('t')
plt.ylabel('W(t)')
plt.show()
```

### Ans 2 and 3:-

```python
A = [1,2,3]
tau = [1,2,3]
t = np.linspace(0,2,100)
W1 = A[0]*(np.exp( -t/tau[0]) - 1 + t/tau[0])
W2 = A[1]*(np.exp( -t/tau[1]) - 1 + t/tau[1])
W3 = A[2]*(np.exp( -t/tau[2]) - 1 + t/tau[2])
plt.plot(t,W1, 'r-')
plt.plot(t,W2, 'b--')
plt.plot(t,W3, 'g-.')
plt.xlabel('t')
plt.ylabel('W(t)')
plt.legend(['W1 (A = 1, tau = 1)','W2 (A = 2, tau = 2)','W3 (A = 3, tau = 3)'])
plt.show()
```

### Ans 4:-

Figure out if the data is given as w and t or t and w - and complete the exercise accordingly

```python
import pandas as pd
import scipy as sc
from scipy.optimize import curve_fit

#define the non-linear function
def W(t,A,tau):
    return A*(np.exp( -t/tau) - 1 + t/tau)

#read the data
data = pd.read_csv('wt.csv',header = None)
xdata = data[1]
ydata = data[0]

#Get the best fit value of A and tau based on the experimental data
popt, pcov = sc.optimize.curve_fit(W,xdata,ydata)


#plot the experimental data with the fitted curve
plt.scatter(xdata,ydata,label = 'Data')
t = np.linspace(0,0.4075,100)
plt.plot(t,W(t,*popt), 'r', label = 'Fitted curve: A = %5.3f, tau = %5.3f' %tuple(popt))
plt.xlabel('t')
plt.ylabel('W(t)')
plt.legend()
plt.show()
```
